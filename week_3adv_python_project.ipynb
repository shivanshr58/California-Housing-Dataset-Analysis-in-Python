# Introduction to the dataset

The dataset contains aggregated data regarding each district in California. It provides information regarding the demography (income, population, house occupancy) in the districts, the location of the districts (latitude, longitude), and general information regarding the house in the districts (number of rooms, number of bedrooms, age of the house).

## Data Description

1. **Longitude**: A measure of how far west a house is; a more negative value is farther west. The values range from -124.3 to -114.3.

2. **Latitude**: A measure of how far north a house is; a higher value is farther north. The values range from 32.5 to 42.5.

3. **Housing Median Age**: Median age of a house within a block; a lower number is a newer building. The values range from 1.0 to 52.0.

4. **Total Rooms**: Total number of rooms within a block. The values range from 2.0 to 37937.0.

5. **Total Bedrooms**: Total number of bedrooms within a block. The values range from 1.0 to 6445.0.

6. **Population**: Total number of people residing within a block. The values range from 3.0 to 35682.0.

7. **Households**: Total number of households, a group of people residing within a home unit, for a block. The values range from 1.0 to 6082.0.

8. **Median Income**: Median income for households within a block of houses (measured in tens of thousands of US Dollars). The values range from 0.5 to 15.0.

9. **Median House Value**: Median house value for households within a block (measured in US Dollars). The values range from 14999.0 to 500001.01.

# Distribution of Median Income

```python
# Sturges' Rule

import math
num_data_points = len(cal_housing_data['housing_median_age']) # calculate the total number of data points

num_bins = round(1 + math.log2(num_data_points))  # calculate the total number of bins
```

```python
# plotting a histogram with matplotlib

plt.hist(cal_housing_data['median_income'],bins = num_bins, color = 'Turquoise',edgecolor = 'black')
plt.title(label="Median income distribution")
plt.xlabel("Income in tens of thousands of US Dollars")
sns.set_style("white")
plt.grid(c = 'gray', ls = 'dashed', lw = 0.5)
plt.show()
```

The histogram shows that the data for Median Income is positively skewed,and the amount of skewness can be calculated with pandas .skew() method.

Knowing the amount of skewness in case of skewed data can help to ascertain which method (Mean or Median) would best represent the central tendency of the data.

# write mode

```python
# checking for outliers

sns.boxplot(cal_housing_data['median_income'])
plt.title('Spread of Median Income & Outliers')
plt.xlabel('Median Income')
plt.show()
```

```python
# calculating the skew

cal_housing_data['median_income'].skew()
```

# Conclusion:

The amount of skew and the box plot together suggest that the data is highly right skewed(>1) and that there are many outliers.

This also means Median would not best represent central tendency here as the presence of high values will inflate the mean value.

As "median_income" has a lot of outliers, and the data is highly right skewed, median would be a better measurement of central tendency than mean for this data.

```python
# calculating the median

print('Median of Median Income --> ',round(cal_housing_data['median_income'].median(),2))
```

# Q6.
 Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

# Writing the get_median function

```python
#defining the get median function,which will take the data set and the columns to calculate mean

def get_median(data,*columns):
  import pandas as pd
  if not columns:
    return data.median()
  else:
    return data[list(columns)].median()

"""The function first checks if any columns have been specified.
If no columns are specified (if not columns),
it calculates and returns the median of the entire DataFrame using data.median().
If one or more columns are specified, it subsets the DataFrame
to only include those columns using data[list(columns)],
and then calculates and returns the median of the subsetted DataFrame."""
```

```python
#function working as expected

get_median(cal_housing_data,'longitude','total_bedrooms','housing_median_age')
```

# Short hand function with lambda


```python
# writing the function with lambda and assigning it to get_median_2

get_median_2 = lambda x,*y: x.median() if not y else x[list(y)].median()
```

```python
#function working as expected

get_median_2(cal_housing_data,'longitude','total_bedrooms','housing_median_age')
```

# Q10.

```python
#calculating maximum bedrooms for defining bins
max_bedrooms = cal_housing_data['total_bedrooms'].max()
```
